```python
# before using this library install by "pip install pandas-profiling"
# Importing the libraries
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
# importing the titanic dataset 
df= pd.read_csv('../input/titanic/train.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
profile = ProfileReport(df, title='Pandas Profiling Report')
```

```python
# there are 2 possible ways to visualise the report
# one is with in the jupyter notebook
profile.to_widgets()
```

```python
# second method is to generate the report in html file
profile.to_file("output.html")
```
